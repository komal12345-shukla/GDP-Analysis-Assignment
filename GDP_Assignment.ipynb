```python
li =[]
for location in all_files:
    df = pd.read_csv(location, index_col = None, header = 0,encoding='ANSI')
    df = df[['Item','2014-15']]
    df = df.set_index('Item')
    df = df.T
    df['States'] = location.split('\\')[-1].split('.')[0].split('-')[-3]
    li.append(df)

frame  = pd.concat(li, axis =0, ignore_index = True)
frame
```

```python
# Dropping UT from dataset as these are not states
frame = frame[frame.States != 'Delhi']
frame = frame[frame.States != 'Chandigarh']
frame = frame[frame.States != 'Puducherry']
```

```python
#Setting 'States' as the index of the DF
frame.set_index('States', inplace = True)

#Setting the data type of the DF to 'float64'
frame = frame.astype('float64')

frame.head()
```

### Categorizing the states based on GDP per capita.

```python
Quantile = frame['Per Capita GSDP (Rs.)'].quantile([.2,.5,.85,1])
Quantile
```

```python
for row in frame.itertuples():
    if frame.loc[row.Index,'Per Capita GSDP (Rs.)'] > Quantile[0.85] and frame.loc[row.Index,'Per Capita GSDP (Rs.)'] <= Quantile[1.00]:
        frame.loc[row.Index,'Category'] = 'C1'
        
    elif frame.loc[row.Index,'Per Capita GSDP (Rs.)'] > Quantile[0.50] and frame.loc[row.Index,'Per Capita GSDP (Rs.)'] <= Quantile[0.85]:
        frame.loc[row.Index,'Category'] = 'C2'
    
    elif frame.loc[row.Index,'Per Capita GSDP (Rs.)'] > Quantile[0.20] and frame.loc[row.Index,'Per Capita GSDP (Rs.)'] <= Quantile[0.50]: 
        frame.loc[row.Index,'Category'] = 'C3'
        
    elif frame.loc[row.Index,'Per Capita GSDP (Rs.)'] < Quantile[0.20]:
        frame.loc[row.Index,'Category'] = 'C4'
       
    
frame = frame.sort_values("Per Capita GSDP (Rs.)", ascending = False)
frame
```

```python
States_Categories = frame[['Agriculture, forestry and fishing','Mining and quarrying','Manufacturing',
                                      'Electricity, gas, water supply & other utility services','Construction',
                                      'Trade, repair, hotels and restaurants',
                                      'Transport, storage, communication & services related to broadcasting',
                                      'Financial services','Real estate, ownership of dwelling & professional services',
                                      'Public administration','Other services','Gross State Domestic Product',
                                      'Per Capita GSDP (Rs.)','Category']]
States_Categories = States_Categories.groupby(['Category'])
C1 = States_Categories.get_group('C1').T
C2 = States_Categories.get_group('C2').T
C3 = States_Categories.get_group('C3').T
C4 = States_Categories.get_group('C4').T
```

```python
C1 = C1.drop(['Category'],axis = 0)
C2 = C2.drop(['Category'],axis = 0)
C3 = C3.drop(['Category'],axis = 0)
C4 = C4.drop(['Category'],axis = 0)
```

```python
C1['Total Sub-Sector'] = C1.sum(axis = 1)
C1 = C1.astype('float64')
C2['Total Sub-Sector'] = C2.sum(axis = 1)
C2 = C2.astype('float64')
C3['Total Sub-Sector'] = C3.sum(axis = 1)
C3 = C3.astype('float64')
C4['Total Sub-Sector'] = C4.sum(axis = 1)
C4_States = C4.astype('float64')
```

```python
gdp_c1 = C1.iloc[-2,4]
C1['% Growth'] = (C1['Total Sub-Sector'] /gdp_c1)*100

gdp_c2 = C2.iloc[-2,9]
C2['% Growth'] = (C2['Total Sub-Sector'] / gdp_c2)*100

gdp_c3 = C3.iloc[-2,8]
C3['% Growth'] = (C3['Total Sub-Sector'] / gdp_c3)*100

gdp_c4 = C4.iloc[-2,6]
C4['% Growth'] = (C4['Total Sub-Sector'] / gdp_c4)*100
```

```python
cumulative_C1 = C1.drop(['Gross State Domestic Product','Per Capita GSDP (Rs.)'], axis=0)
cumulative_C1 = cumulative_C1.sort_values('% Growth', ascending=False)
cumulative_C1['Cumulative'] = cumulative_C1['% Growth'].cumsum()

cumulative_C2 = C2.drop(['Gross State Domestic Product','Per Capita GSDP (Rs.)'], axis=0)
cumulative_C2 = cumulative_C2.sort_values('% Growth', ascending=False)
cumulative_C2['Cumulative'] = cumulative_C2['% Growth'].cumsum()

cumulative_C3 = C3.drop(['Gross State Domestic Product','Per Capita GSDP (Rs.)'], axis=0)
cumulative_C3 = cumulative_C3.sort_values('% Growth', ascending=False)
cumulative_C3['Cumulative'] = cumulative_C3['% Growth'].cumsum()

cumulative_C4 = C4.drop(['Gross State Domestic Product','Per Capita GSDP (Rs.)'], axis=0)
cumulative_C4 = cumulative_C4.sort_values('% Growth', ascending=False)
cumulative_C4['Cumulative'] = cumulative_C4['% Growth'].cumsum()
```

```python
top_sector_c1 = cumulative_C1[cumulative_C1.Cumulative < 85]
top_sector_c1
```

```python
top_sector_c2 = cumulative_C2[cumulative_C2.Cumulative < 85]
top_sector_c2
```

```python
top_sector_c3 = cumulative_C3[cumulative_C3.Cumulative < 85]
top_sector_c3
```

```python
top_sector_c4 = cumulative_C4[cumulative_C4.Cumulative < 85]
top_sector_c4
```
